```python
#PatternCount
def PatternCount(Text, Pattern):
    count = 0
    for i in range(len(Text)-len(Pattern)+1):
        if Text[i:i+len(Pattern)] == Pattern:
            count +=1
    return count
```

```python
#FrequentWords

def FrequenWords(Text,k):
    words = []
    Count = {} # key:index, values:start_point
    for i in range(len(Text)-k+1):
        word = Text[i:i+k]
        Count[i] = PatternCount(Text,word)
    max_count = max(Count.values()) 
    #print(Count)
    for i in Count:
        if Count[i] == max_count:
            words.append(Text[i:i+k])
    return set(words)
                   
```

```python
#The Frequency Array
SymbolToNumber = {'A':0, 'C':1, 'G':2, 'T':3}
NumberToSymbol = {0:'A', 1:'C', 2:'G', 3:'T'}
def Pattern2Number(Pattern):
    if len(Pattern)==1:
        return SymbolToNumber[Pattern]
    Symbol = Pattern[-1]
    pattern = Pattern[:-1]
    return 4 * Pattern2Number(pattern) + SymbolToNumber[Symbol]

def Number2Pattern(index,k):
    index = int(index)
    k = int(index)
    if k ==1:
        return NumberToSymbol[k]
    else:
        p = index//4
        rest = index%4
        result = NumberToSymbol[rest]
        return Number2Pattern(p,k-1) + result
    
```

```python
#FindingFrequentWordsBySorting
def FindingFrequentWordBySorting(Text,k):
    FrequentPatterns = []
    k = int(k)
    Index = {}
    Count = {}
    for i in range(len(Text)-k+1):
        Pattern = Text[i:i+k]
        Index[i] = Pattern2Number(Pattern)
        Count[i] = 1
    Sorted_index = sorted(Index)
    
    for i in range(1,len(Text)-k+1):
        if Sorted_index[i] == Sorted_index[i-1]:
            Count[i] = Count[i-1]+1
    maxCount = max(Count.values())
    
    for i in range(len(Text)-k+1):
        if Count[i] == maxCount:
            Pattern = Number2Pattern(Sorted_index[i],k)
            FrequentPatterns.append(Pattern)
    return FrequentPatterns
    
```
